## convert to gray image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


# image path
filepath = "images/DBV_Demo.png"
color = cv2.imread(filepath)

# convert into gray-scale
gray = cv2.cvtColor(color, cv2.COLOR_BGR2GRAY)

# gray image show
plt.imshow(gray, cmap='gray', vmin=0, vmax=255)
plt.axis('off')
plt.show()
```

## save processed image

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


# image path
filepath = "images/DBV_Demo.png"
color = cv2.imread(filepath)

# convert into gray-scale
gray = cv2.cvtColor(color, cv2.COLOR_BGR2GRAY)
cv2.imwrite('images/gray_image.png', gray)

import os
if os.path.exists('images/gray_image.png'):
    print('processed image has been saved')
```

## image compression


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


filename = "images/coins.png"
gray = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)
plt.figure()  
plt.imshow(gray, cmap="gray", vmin=0, vmax=255)
plt.title('original image')
plt.axis('off')
plt.show()


# compress into jpg
cv2.imwrite("images/compressed.jpg", gray, [int(cv2.IMWRITE_JPEG_QUALITY), 30])
compressed = cv2.imread("images/compressed.jpg", cv2.IMREAD_GRAYSCALE)
plt.figure()
plt.imshow(compressed, cmap="gray", vmin=0, vmax=255)
plt.title('compressed image')
plt.axis('off')
plt.show()


# Differenz zum Originalbild
difference = np.float32(gray) - np.float32(compressed)
plt.figure()
plt.imshow(difference, cmap="gray")
plt.title('difference')
plt.axis('off')
plt.show()
```

## histogram

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import glob


#%% Bild einlesen und in Graustufen umwandeln
filename = "images/DBV_Demo.png"
gray = cv2.imread(filename, cv2.IMREAD_GRAYSCALE)


#%% Bild anzeigen
plt.figure()
plt.imshow(gray, cmap="gray")
plt.axis('off')
plt.show()


#%% Histogramm berechnen und plotten
hist = cv2.calcHist(images=[gray], channels=[0], mask=None, histSize=[256], ranges=[0,256])


plt.figure()
plt.plot(hist)
plt.xlabel("Grauwert")
plt.ylabel("Hufigkeit")
plt.show()
```

## binarization

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np


# read gray image 
gray = cv2.imread("images/Fingerabdruck.png", cv2.IMREAD_GRAYSCALE)
plt.figure()
plt.imshow(gray, cmap="gray", vmin=0, vmax=255)
plt.title('original image')
plt.axis('off')
plt.show()


# calculate histogram of gray scale
hist = cv2.calcHist(images=[gray],channels=[0],mask=None,histSize=[256],ranges=[0,256])
plt.figure()
plt.plot(hist)
plt.xlabel("Grauwert")
plt.ylabel("Hufigkeit")
plt.title('gray scale histogram')
plt.show()

# pick threshold and binarization
threshold = 110
plt.figure()
mask = gray > threshold       
plt.imshow(mask, cmap="gray", vmin=0, vmax=1)
plt.title('binarized image')
plt.axis('off')
plt.show()
```

## Gamma Correction

${img1} = 255 * c * \frac{img}{255}^{gamma} $

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import glob


# read original image 
img = cv2.imread("images/Einstein_LR.png", cv2.IMREAD_GRAYSCALE) # gleich Graubild lesen
plt.figure()
plt.imshow(img, cmap="gray", vmin=0, vmax=255)
plt.title('original image')
plt.axis('off')
plt.show()


# Gamma Correction and corrected image
c = 1.5
gamma = 2
adjusted = c * (img/255) ** gamma * 255

plt.figure()
plt.imshow(adjusted, cmap = 'gray', vmin=0, vmax=255)
plt.title('corrected image')
plt.axis('off')
plt.show()


# plot transformation function
x = np.arange(0,256) 
y = c * (x/255) ** gamma * 255 

plt.plot(x, y, label="c={}, ={}".format(c, gamma))
plt.xlabel('Grauwert '+r'$\mathrm{I}_{inp}$')
plt.ylabel('Grauwert '+r'$\mathrm{I}_{out}$')
plt.text(0,300, r'$\mathrm{I}_{out}=c \cdot {{I}_{inp}}^{\gamma}$', size=16)
plt.title('transformation function')
plt.legend(loc='best')
plt.show()
```

## histogram contrast

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

#%% Histogrammausgleich
# read image
picture = cv2.imread('images/Einstein_LR.png',cv2.IMREAD_GRAYSCALE)

# apply histogram equalization
equalized = cv2.equalizeHist(picture)

plt.imshow(equalized, cmap = 'gray')
plt.show()

#%% Contrast Limited Adaptive Histogram Equalization (CLAHE)
# Adaptive histogram equalization -> segmentation in tiles(=8*8)
# Contrast limited -> histogram buckets above limit(=40) are clipped

img = cv2.imread('images/Einstein_LR.png',cv2.IMREAD_GRAYSCALE)

# create a Contrast Limited Adaptive Histogram Equalization (CLAHE) object (Arguments are optional)
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
img_adj = clahe.apply(img)

plt.figure()
plt.imshow(img_adj, cmap = 'gray')
plt.show()
```
